```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import keras
```

```python
(xtrain,ytrain), (xtest,ytest)=keras.datasets.fashion_mnist.load_data()
```

```python
xtrain.shape
```

```python
ytrain.shape
```

```python
xtrain[0]
```

```python
ytrain[0]
```

```python
class_labels = [
    "T-shirt/top",
    "Trouser",
    "Pullover",
    "Dress",
    "Coat",
    "Sandal",
    "Shirt",
    "Sneaker",
    "Bag",
    "Ankle boot"
]
```

```python
plt.imshow(xtrain[0],cmap="Greys")
```

```python
xtrain=np.expand_dims(xtrain,-1)
```

```python
xtest=np.expand_dims(xtest,-1)
```

```python
xtrain[0]
```

```python
xtrain=xtrain/255
xtest=xtest/255
```

```python
xtrain[0]
```

```python
from sklearn.model_selection import train_test_split
xtrain,xval,ytrain,yval=train_test_split(xtrain,ytrain,test_size=0.2,random_state=1)
```

```python
xtrain.shape
```

```python
xval.shape
```

```python
model=keras.models.Sequential([
    keras.layers.Conv2D(filters=32,kernel_size=3,activation="relu",input_shape=[28,28,1]),
    keras.layers.MaxPool2D(pool_size=(2,2)),
    keras.layers.Conv2D(filters=32,kernel_size=3,activation="relu"),
    keras.layers.MaxPool2D(pool_size=(2,2)),
    keras.layers.Flatten(),
    keras.layers.Dense(units=128,activation="relu"),
    keras.layers.Dropout(0.25),
    keras.layers.Dense(units=256,activation="relu"),
    keras.layers.Dropout(0.25),
    keras.layers.Dense(units=128,activation="relu"),
    keras.layers.Dense(units=10,activation="softmax"),

])
```

```python
model.summary()
```

```python
model.compile(optimizer="adam",loss="sparse_categorical_crossentropy",metrics=['accuracy'])
```

```python
model.fit(xtrain,ytrain,epochs=20,batch_size=500,verbose=1,validation_data=(xval,yval))
```

```python
model.predict(np.expand_dims(xtest[0],axis=0)).round(2)
```

```python
np.argmax(model.predict(np.expand_dims(xtest[0],axis=0)).round(2))
```

```python
ytest[0]
```

```python
ypred=model.predict(xtest).round(2)
ypred
```

```python
model.evaluate(xtest,ytest)
```

```python
from sklearn.metrics import confusion_matrix,classification_report
ypredlabel= [np.argmax(i) for i in ypred]
cm=confusion_matrix(ytest,ypredlabel)
cm
```

```python
sns.heatmap(cm,annot=True,fmt='d',xticklabels=class_labels,yticklabels=class_labels)
```

```python
cr=classification_report(ytest,ypredlabel,target_names=class_labels)
print(cr)
```
